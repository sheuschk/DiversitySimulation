#### An objective function for a team

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(0)
```

```python
x_values = np.unique(rng.integers(0,100,20))
y_values = rng.normal(.5, 0.2, x_values.shape)
plt.scatter(x_values, y_values)
```

```python
from scipy import interpolate
```

```python
f = interpolate.interp1d(x_values, y_values)
```

```python
x = np.arange(x_values.min(), x_values.max())
plt.plot(x, f(x))
plt.scatter(x_values, y_values)
```

```python
f = interpolate.CubicSpline(x_values, y_values)
plt.plot(x, f(x))
```

```python
f = interpolate.PchipInterpolator(x_values, y_values)
plt.plot(x, f(x))
```

```python
f = interpolate.make_interp_spline(x_values, y_values, k=2) # k=3 is cubic
plt.plot(x, f(x))
```

```python
f = interpolate.interp1d(x_values, y_values,  kind="nearest") # kind=’nearest’, ‘previous’, ‘next’
plt.plot(x, f(x))
```

##### Individual competencies


```python
f = interpolate.CubicSpline(x_values, y_values)

for i in range(3):
    x_start = rng.integers(x_values.min(), x_values.max()-15, 1)
    # Option 1: Draw from interpolized with normal disribution, having a cloe to perfect dist for an area
    # Option 2: Sample the original x_values in this area and interpolate
    # Option 3: Just give the original values of interpolation 
    ind_x = np.linspace(x_start, x_start+10, 5)
    ind_y = f(ind_x)
    plt.plot(ind_x, ind_y)
plt.scatter(x_values, y_values)
plt.plot(np.linspace(x_values.min(), x_values.max(), 100), f(np.linspace(x_values.min(), x_values.max(), 100)), alpha=0.3)
```
